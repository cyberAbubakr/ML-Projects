# Error Analysis & Model Explanation

## 1. Residual Analysis

Residual analysis was performed to understand prediction errors produced by the champion model.

The residual is calculated as:

\[
Residual = Actual - Predicted
\]

Residuals help identify:

- Whether the model systematically overpredicts or underpredicts
- The spread of prediction errors
- Areas where the model has difficulty learning patterns

A well-performing regression model should have residuals distributed around zero without strong patterns.

---

# 2. Actual vs Predicted Analysis

The relationship between actual and predicted values was analyzed using prediction plots.

Interpretation:

- Points close to the diagonal line represent accurate predictions.
- Points far away from the line indicate larger prediction errors.
- Large deviations usually occur in unusual trips or rare traffic conditions.

---

# 3. Large Error Cases

The largest prediction errors were investigated to understand difficult prediction scenarios.

Common causes of large errors include:

- Unexpected traffic congestion
- Unusual trip distances
- Rare pickup/drop-off location combinations
- Airport delays
- Events affecting travel time

These cases represent situations where historical features cannot fully explain trip duration variation.

---

# 4. Feature-Based Error Analysis

Prediction errors were analyzed across important trip characteristics.

## Trip Distance

Longer trips generally have higher uncertainty because:

- Traffic conditions vary significantly
- Route selection affects travel time
- Delays accumulate over longer distances

---

## Rush Hour Trips

Rush hour periods introduce additional uncertainty because:

- Traffic congestion changes rapidly
- Travel speeds become less predictable
- Similar trips may have different durations

---

## Airport Trips

Airport trips may have higher prediction difficulty because:

- Passenger waiting times vary
- Airport traffic patterns are different
- Pickup and drop-off restrictions affect travel time

---

# 5. Champion Model Explanation

The selected champion model is:

## Tuned Random Forest Regressor

The model achieved:

| Metric | Score |
|---|---:|
| MAE | 0.195348 |
| RMSE | 0.261017 |
| Median AE | 0.154788 |
| R² Score | 0.848577 |

Random Forest performed best because it can capture complex nonlinear relationships between:

- Trip distance
- Pickup and drop-off locations
- Time features
- Rush hour patterns
- Airport indicators
- Passenger information

Tree-based models are better suited than linear models for this problem because taxi duration depends on multiple interacting factors.

---

# 6. Model Limitations

Although the champion model achieved strong performance, some prediction errors remain due to missing external information.

Potential sources of error:

- Real-time traffic conditions
- Weather conditions
- Road closures
- Driver behavior
- Unexpected passenger delays
- Special events

---
