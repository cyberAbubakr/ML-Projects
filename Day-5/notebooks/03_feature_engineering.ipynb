## Fix Airport Fee Columns

````python
if "Airport_fee" in featured_df.columns:

    if "airport_fee" not in featured_df.columns:
        featured_df["airport_fee"] = (
            featured_df["Airport_fee"]
        )

    else:
        featured_df["airport_fee"] = (
            featured_df["airport_fee"]
            .fillna(featured_df["Airport_fee"])
        )

    featured_df["airport_fee"] = (
        featured_df["airport_fee"]
        .fillna(0)
    )

    featured_df.drop(
        columns=["Airport_fee"],
        inplace=True
    )
````

## Convert String Columns

````python
if "store_and_fwd_flag" in featured_df.columns:

    if featured_df["store_and_fwd_flag"].dtype == "object":

        featured_df["store_and_fwd_flag"] = (
            featured_df["store_and_fwd_flag"]
            .map({
                "N": 0,
                "Y": 1
            })
            .fillna(0)
            .astype("int8")
        )
````

## Remove Unnecessary Features

````python
featured_df.drop(
    columns=["route_id"],
    errors="ignore",
    inplace=True
)
````

## Optimize Datatypes

````python
binary_cols = [
    "is_weekend",
    "morning_rush",
    "evening_rush",
    "night_trip",
    "same_location"
]

for col in binary_cols:
    if col in featured_df.columns:
        featured_df[col] = (
            featured_df[col]
            .astype("int8")
        )
````

# Feature Engineering Pipeline Summary

## Overview

The feature engineering pipeline was successfully executed on the cleaned NYC Taxi dataset to transform raw trip information into a machine-learning-ready dataset. The objective was to create meaningful predictive features that capture temporal patterns, trip characteristics, and spatial relationships affecting taxi trip duration.

## Dataset Transformation

The pipeline maintained the complete dataset size of:

* **Rows:** 18,378,485 taxi trips
* **Original Features:** 14 columns
* **Final Engineered Features:** 27 columns

No trip records were removed during feature generation, ensuring complete data availability for model training and evaluation.

## Feature Engineering Applied

### 1. Time-Based Features

Created features to capture travel behavior changes over different time periods:

* Pickup hour
* Pickup day
* Day of week
* Month
* Week number
* Weekend indicator

These features allow models to learn daily and seasonal travel patterns.

### 2. Rush Hour Features

Added traffic-related indicators:

* Morning rush hour
* Evening rush hour
* Night trips

These features help capture variations in trip duration caused by congestion patterns.

### 3. Cyclical Time Encoding

Applied cyclical transformations to represent repeating time patterns:

* Hour sine transformation
* Hour cosine transformation
* Month sine transformation
* Month cosine transformation

This prevents models from treating time values as simple linear numbers.

### 4. Location and Route Features

Created spatial relationship features:

* Pickup and drop-off location relationships
* Same location indicator

Removed high-cardinality `route_id` feature because raw route identifiers provide limited generalization and increase model complexity.

### 5. Distance-Based Features

Created additional trip behavior features:

* Distance per passenger
* Trip distance relationships

These features provide information about trip efficiency and passenger load.

### 6. Data Cleaning and Optimization

Final preprocessing steps included:

* Combined duplicate airport fee columns
* Converted `store_and_fwd_flag` into binary format
* Removed unnecessary columns
* Optimized binary feature memory usage
* Verified no remaining string columns
* Verified missing values

## Validation Results

Final validation confirmed:

* ✅ No object/string columns remain
* ✅ No duplicate feature columns
* ✅ No missing values requiring treatment
* ✅ Dataset is compatible with machine learning algorithms
