# 02 - Data Preprocessing

## Objective

The goal of this notebook is to execute and validate the preprocessing pipeline implemented in `src/data/build_dataset.py`.

The preprocessing stage transforms the raw NYC Yellow Taxi dataset into a clean, machine-learning-ready dataset by creating the target variable, removing invalid records, handling data quality issues, and eliminating target leakage.

---

# Tasks

## 1. Execute the Preprocessing Pipeline
- Load the raw dataset
- Create the target variable (`trip_duration_minutes`)
- Create the log-transformed target (`log_trip_duration`)

---

## 2. Clean Invalid Records
- Remove corrupted timestamps (e.g., year 2001)
- Remove trips where dropoff time is earlier than pickup time
- Remove trips with zero or negative duration
- Remove trips longer than the selected threshold
- Remove trips with zero or negative distance
- Remove invalid passenger counts

---

## 3. Handle Missing Values
- Inspect missing values after cleaning
- Handle missing values where required
- Verify that the dataset is suitable for modeling

---

## 4. Remove Target Leakage
Remove columns that are only available after the trip has ended, including fare and payment-related information, to ensure realistic model training.

---

## 5. Validate the Cleaned Dataset
- Verify dataset shape
- Verify target variable
- Verify passenger counts
- Verify trip duration
- Verify timestamps
- Confirm leakage columns have been removed

---

## 6. Save Processed Dataset
Save the cleaned dataset to:

`data/interim/cleaned_data.parquet`

This dataset will be used in the Feature Engineering stage.

---

# Expected Outcome

At the end of this notebook, we should have:

- A clean dataset
- A valid target variable
- No corrupted timestamps
- No invalid passenger counts
- No impossible trip durations
- No target leakage
- A dataset ready for feature engineering

---



```python
# ==========================================================
# Save Processed Dataset
# ==========================================================

from pathlib import Path

# Create the output directory if it doesn't exist
output_dir = Path("../data/interim")
output_dir.mkdir(parents=True, exist_ok=True)

# Define output file path
output_file = output_dir / "cleaned_data.parquet"

# Save the cleaned DataFrame
cleaned_df.to_parquet(output_file, index=False)

print("=" * 60)
print("✅ Processed dataset saved successfully!")
print(f"📁 Location : {output_file.resolve()}")
print(f"📊 Shape    : {cleaned_df.shape}")
print("=" * 60)
```
