# Data Type Validation

## Objective

Data type validation ensures that every feature has the correct data type before preprocessing and model development.

Using appropriate data types is important because machine learning algorithms expect numerical and categorical features to be represented correctly.

This analysis helps identify:

- Numerical features
- Categorical features
- Incorrect data types
- Columns that may require type conversion

---

## Important Note

This section only validates data types.

No data type conversions are performed in this notebook.

Any required type conversions will be carried out during the preprocessing stage.

```python
df.dtypes
```

```python
dtype_summary = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.astype(str)
})

dtype_summary
```

```python
df.dtypes.value_counts()
```

```python
numerical_features = df.select_dtypes(include=["number"]).columns.tolist()

categorical_features = df.select_dtypes(include=["object", "string"]).columns.tolist()

print("Numerical Features:")
print(numerical_features)

print("\nCategorical Features:")
print(categorical_features)
```

```python
feature_summary = pd.DataFrame({
    "Feature": df.columns,
    "Type": [
        "Numerical" if col in numerical_features else "Categorical"
        for col in df.columns
    ],
    "Data Type": df.dtypes.astype(str).values
})

feature_summary
```
