## Correlation Analysis

Correlation analysis measures the relationship between numerical features.

The correlation coefficient ranges from -1 to +1:

- +1 indicates a strong positive relationship.
- 0 indicates no linear relationship.
- -1 indicates a strong negative relationship.

Correlation analysis helps identify:

- Related features
- Possible redundant variables
- Important relationships between numerical features

No features are removed during this analysis.

## Distribution Analysis

Distribution analysis helps us understand how numerical values are spread across the dataset.

It helps identify:

- Data concentration
- Skewness
- Spread of values
- Possible extreme values

In this analysis, we will examine:

- Age distribution
- Hours per week distribution
- Capital gain distribution

No data modification will be performed.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="age",
    bins=30,
    kde=True
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

## Age Distribution Findings

The age feature shows a slightly right-skewed distribution.

Observations:

- Most individuals are concentrated between approximately 20 and 45 years old.
- The highest frequency occurs around the early-to-mid 30s.
- After age 45, the number of individuals gradually decreases.
- The distribution extends toward older ages, reaching approximately 90 years.

Statistical implication:

- The dataset mainly represents working-age adults.
- The long tail toward higher ages indicates fewer elderly individuals.
- The distribution is not perfectly symmetric.

No changes are applied during this analysis.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="hours_per_week",
    bins=40,
    kde=True
)

plt.title("Hours Per Week Distribution")
plt.xlabel("Hours Worked Per Week")
plt.ylabel("Frequency")

plt.show()
```

## Hours Per Week Distribution Findings

The hours_per_week feature shows a concentrated distribution with multiple peaks.

Observations:

- The largest peak occurs at approximately 40 hours per week.
- This indicates that a standard full-time working schedule is most common.
- Additional peaks appear around:
    - 20 hours
    - 45 hours
    - 50 hours
    - 60 hours

These patterns suggest the presence of:

- Part-time workers
- Full-time workers
- Individuals working overtime

Statistical implication:

- The distribution is not perfectly normal.
- The feature contains common working patterns rather than random values.

No modifications are applied during this analysis.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="capital_gain",
    bins=50,
    kde=True
)

plt.title("Capital Gain Distribution")
plt.xlabel("Capital Gain")
plt.ylabel("Frequency")

plt.show()
```

## Capital Gain Distribution Findings

The capital gain feature shows a highly right-skewed distribution.

Observations:

- The majority of individuals have a capital gain value of 0.
- A very small number of individuals have extremely high capital gains.
- The distribution contains a long right tail extending toward the maximum value.
- The extreme values indicate that capital gain is not normally distributed.

Statistical implication:

- The feature has high skewness.
- Mean and median are significantly different.
- Further investigation may be required during data quality analysis.

No changes are applied during this analysis.
